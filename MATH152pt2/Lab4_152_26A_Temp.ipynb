```python
# Enter your code here
x = sp.symbols('x')
f = sp.exp(-x/2)* sp.cos(sp.pi*x)
f_prime = sp.diff(f,x)
p = plot(f, (x, sp.Rational(-3,4), sp.Rational(3,4)),
         title="f(x)",
         xlabel="x", ylabel="f(x)", show=False)
p.show()

print("f'(x) =", f_prime)
critical_points = []
cp  = sp.nsolve(f_prime, x, -.1)
critical_points.append(cp)
print("Critical point =", cp)


```

#### 1c

```python
# Enter your code here
a = sp.Rational(-3,4)
b = sp.Rational(3,4)
n = 20
dx = (b-a)/n

endpoints = [a + i*dx for i in range(n+1)]
Lw_points = []
U_points = []
for i in range(n):
    x_left = endpoints[i]
    x_right = endpoints[i+1]
    cands = [x_left, x_right]
    for cp in critical_points:
        if x_left < cp < x_right:
            cands.append(cp)
    vals = [(xi, float(f.subs(x, xi))) for xi in cands]
    min_pts = min(vals, key=lambda t: t[1])
    max_pts = max(vals, key=lambda t: t[1])
    Lw_points.append(min_pts[0])   
    U_points.append(max_pts[0])    

dx_float_version = float(dx)
Lw20 = sum(float(f.subs(x, xi)) * dx_float_version for xi in Lw_points)
U20  = sum(float(f.subs(x, xi)) * dx_float_version for xi in U_points)

print("Lw20 =", Lw20)
print("U20 =", U20)
```

```python
# Enter your code here

x = sp.symbols('x')
numer = -3*x**3 + 5*x + 9
denom = x**2 - 4*x + 3
f = numer / denom
lim_pos = sp.limit(f,x,sp.oo)
lim_neg = sp.limit(f,x,-sp.oo)
print("lim x-> +inf f(x) =", lim_pos)
print("lim x-> -inf f(x) =", lim_neg)
```

#### 2b

```python
# Enter your code here
split = sp.apart(f, x)
print("Partial fraction decomposition =", split)
q =  -3*x - 12
g = split - q
g = sp.simplify(g)


```

#### 2c
