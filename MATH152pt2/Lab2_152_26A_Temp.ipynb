```python
#start code here
r1, r2 = sp.symbols('r1 r2', positive=True)
degrees = sp.pi/180

y_circle = sp.sqrt(R**2 - x**2)

y1 = x*sp.tan(degrees*r1)
y2 = x*sp.tan(degrees*r2)

x1 = R*sp.cos(degrees*r1)
x2 = R*sp.cos(degrees*r2)

A1 = sp.integrate(y_circle - y1, (x, 0, x1))
A2 = sp.integrate(y_circle - y2, (x, 0, x2))

A_sector = sp.simplify(A1 - A2)
A_sector


```
