```python
# Enter your code here
y, L = sp.symbols('y L')
rho = sp.Rational(624,10)
A1 = width * (L + sp.Rational(12,5)*y)
A2 = width * (L + 17)
A3 = width * (L + 19)
A4 = width * (L + 21)
W1_L = sp.integrate(rho * A1* (7 - y), (y, 0,sp.Rational(5,2)))
W2_L = sp.integrate(rho * A2 * (7 - y), (y, sp.Rational(5,2), 4))
W3_L = sp.integrate(rho * A3 * (7 - y), (y, 4,sp.Rational(11,2)))
W4_L = sp.integrate(rho * A4 * (7 - y), (y, sp.Rational(11,2), 7))

W_total_L = sp.expand(W1_L + W2_L + W3_L + W4_L)
# max load 1,000,000
sol = sp.solve(sp.Eq(W_total_L, 1000000), L)
for s in sol:
    if s > 0:
        print("Maximum load length = ", s)
print("This is the max length of deep end")
```

```python
import random
import math
# Enter your code here

successes = 0
total = 1_000_000
for i in range(total):
    y = random.uniform(-1, 1)
    x = random.uniform(-1, 1)
    if x**2 + y**2 <= 1:
        successes += 1

pi_estimate = 4 * successes / total
print("Estimated value of pi: ", pi_estimate)

```

#### 3b

```python
# Enter your code here
import random
import math

successes = 0
total = 10_000

for i in range(total):
    x = random.uniform(-math.pi / 2, math.pi / 2)
    y = random.uniform(-math.pi / 2, math.pi / 2)
    r = math.sqrt(x**2 + y**2)
    if r <= math.pi / 2:
        if x * math.tan(r) <= y:
            successes += 1

square_area = math.pi ** 2
area_estimate = square_area * successes / total
print("Estimated area of the region: ", area_estimate)
```
