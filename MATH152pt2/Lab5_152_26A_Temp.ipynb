### Question 1

#### 1a

```python
# Enter your code here
n_vals = np.arange(1,51)

a_a = ((-1)**n_vals) * (1 / np.log(n_vals + 1))
b_a = np.cumsum(a_a) / n_vals

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(n_vals, a_a, 'o-', markersize=4, label=r'$a_n = (-1)^n / \ln(n+1)$')
ax.plot(n_vals, b_a, 's-', markersize=4, label=r'$b_n$ (arithmetic avg)')
ax.axhline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlabel('n')
ax.set_title('Part a)')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()

print("a_n oscillates, and slowly converges to 0 as n increases.")
print("b_n converges to 0 as n increases, but it does so more smoothly")
```

#### 1b

```python
# Enter your code here
a_b = 2 - 1/n_vals
b_b = np.cumsum(a_b) / n_vals

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(n_vals, a_b, 'o-', markersize=4, label=r'$a_n = 2 - 1/n$')
ax.plot(n_vals, b_b, 's-', markersize=4, label=r'$b_n$ (arithmetic avg)', color='tomato')
ax.axhline(2, color='gray', linestyle='--', linewidth=1, label='limit = 2')
ax.set_xlabel('n')
ax.set_title('Part b)')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
print("a_n increases towards 2, so lim a_n = 2")
print("b_n also converges to 2, approaches slowly due to the early terms")
```

#### 1c

```python
# Enter your code here
c_n = np.cumprod(a_b) ** (1 / n_vals)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(n_vals, b_b, 's-', markersize=4, label=r'$b_n$ (arithmetic avg)', color='tomato')
ax.plot(n_vals, c_n, '^-', markersize=4, label=r'$c_n$ (geometric avg)', color='steelblue')
ax.axhline(2, color='gray', linestyle='--', linewidth=1, label='limit = 2')
ax.set_xlabel('n')
ax.set_title('Part c)  —  Arithmetic vs Geometric Average of $a_n = 2 - 1/n$')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()

print("c_n converges to 2, but it lies below b_n for all n, showing that the geometric average is less than or equal to the arithmetic average for this sequence.")

```

```python
# Enter your code here
x , t = sp.symbols('x t')
r = 3
f = r/x
fp = sp.diff(f, x)

V_truncated = sp.pi * sp.integrate(f**2 , (x,1,t))
print("Volume of truncated cone V(t) =", V_truncated)
```

#### 2b

```python
# Enter your code here
V_full = sp.limit(V_truncated, t, sp.oo)
print("Volume of the full cone V =", V_full)
```

#### 2c

```python
# Enter your code here
SA_integrand = 2 * sp.pi * f * sp.sqrt(1 + fp**2)
SA_truncated = sp.integrate(SA_integrand, (x, 1, t))
print("Surface area of truncated cone SA(t) =", SA_truncated)

```

#### 2d

```python
# Enter your code here
SA_full = sp.limit(SA_truncated, t, sp.oo)
print("Surface area of the full cone SA =", SA_full)

```
